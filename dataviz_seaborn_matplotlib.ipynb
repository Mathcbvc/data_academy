Gráfico de Barras : 
é um tipo de plotagem usado para representar dados categóricos com barras retangulares. Cada barra representa uma categoria e a altura da barra representa a quantidade ou frequencia da cateforia. 

O eixo horizontal do grafico de barras mostra as categorias e o eixo vertical mostra a escala de medida dos dados. As barras podem ser vertical ou horizontal, dependendo da preferencia do usuario. 

Os graficos de barras sao comumente usados para comparar a quantidade ou frequencia de diferentes categorias. Eles sao uteis para mostrar diferencas entre grupos e ajudam rapidamente quais categorias tem valores maiores ou menores. 

```python
x1 = [2,4,6,8,10]
y1 = [6,7,8,2,4]

plt.bar(x1, y1, label= ['Barras_1','Barras_2','Barras_3','Barras_4','Barras_5'] , color = ['Blue', 'green','red','pink','orange'])
plt.legend()
plt.show()
```

```python
x2 = [1,3,5,7,9]
y2 = [7,8,2,4,2]

plt.bar(x1, y1, label='barras_1' , color ='green')
plt.bar(x2, y2, label='barras_2' , color ='blue')
plt.legend()
plt.show() #conceito de gramatica dos gráficos. Como estao na mesma area de plotagem, ambos serao plotados na mesma área de plotagem.
```

```python
 #Dados
x = linspace(0,5,10)
y = x **2

#Cria os subplots 
fig, axes = plt.subplots(1,2,figsize =(10,3)) # subplots definidos como 1 linha e 2 colunas 

#Grid padrão - PLOT 1
axes[0].plot(x, x**2, x, x **3, lw= 2)
axes[0].grid(True)

#Grid customizado - PLOT 2
axes[1].plot(x, x**2, x, x **3, lw= 2)
axes[1].grid(color = 'b', alpha = 0.2 , linestyle ='dashed' , linewidth = 0.8)
```

```python
#Histogramas - avaliar a distribuição de uma variável contínua

#Dados 
n = np.random.randn(10000)

#Cria os subplots
fig, axes = plt.subplots(1,2,figsize = (12,4))

#Plot 1
axes[0].hist(n)
axes[0].set_title('Histograma Padrão')
axes[0].set_xlim((min(n), max(n)))

#Plot 2 
axes[1].hist(n , cumulative=True, bins = 50)
axes[1].set_title('Histograma Cumulativo')
axes[1].set_xlim((min(n), max(n)));

```

```python
#importações necessárias
import random
import numpy as np
import warnings
import matplotlib as mat
import matplotlib.pyplot as plt
import pandas as pd

warnings.filterwarnings("ignore")
```

```python
dados = sea.load_dataset("tips")
```

```python
dados.sample(5)
```

```python
#Método joinpoint cria plot de 2 variáveis com gráficos bivariados e univariados
sea.jointplot(data= dados , x = "total_bill", y ="tip", kind= 'reg');
```

```python
#metodo lmplot cria plot com dados e modelos de regressão 
sea.lmplot(data = dados , x ='total_bill', y = 'tip', col='smoker');
```

```python
#criando um dataframe com pandas
df = pd.DataFrame()

#alimentando o DF 
df['idade'] = random.sample(range(20,100),30)
df['peso'] = random.sample(range(55,150),30)

df.sample(5)
```

```python
#lmplot
sea.lmplot(data = df, x ='idade', y='peso', fit_reg = True)#fit_reg é um parâmetro que indica a criação de um modelo de regressão pelo seaborn 
#a área sombreada significa intervalo de confiança - funciona mais ou menos como uma margem de erro
#a linha azul representa a regressão qeu ajuda a explicar o comportamento da relação entre as variáveis de peso X idade
# Nesse caso existe uma correlação positiva. Caso a linha estivesse descendo, seria uma correlação negativa e caso a linha estivesse reta, seria uma correlação neutra (). 
;
```

```python
#kdeplot - graíco de densidade - usado para analisar a distribuição de uma variável
sea.kdeplot(df.idade)
;
```

```python
sea.kdeplot(df.peso);
```

```python
#distplot une o histograma e o gráfico de densidade
sea.distplot(df.peso)
;
```

```python
#Histograma
plt.hist(df.idade, alpha = .3) #utilizando matplotlib pyplot para criar o gráfico de histograma
sea.rugplot(df.idade) # dentro do histograma criado, criou-se o rugplot para demarcar melhor a variável idade e sua distribuição 
;
```

```python
#boxplot - Gráfico a linha acima do gráfico representa o terceiro quartil, a linha que fica mais ou menos ao meio representa a mediana (segundo quartil) e a linha abaixo representa o primeiro quartil. 
sea.boxplot(df.idade, color='m') 
;
```

```python
#Unindo numpy , pandas, matplotlib e seaborn 

#Gerando valores randomicos - usando numpy
np.random.seed(42)
n = 1000
pct_smokers = .2

#variaveis - usando numpy
flag_fumante = np.random.rand(n) < pct_smokers #forçou a criação de um true e false para converter em fumante e nao fumante depois

idade = np.random.normal(40,10,n)
altura = np.random.normal(170,10, n)
peso = np.random.normal(70,10, n)

#Dataframe - usando pandas
dados = pd.DataFrame({'altura': altura.round(decimals=1), 'peso':peso.round(decimals=1), 'idade':idade.round().astype('int32'), 'flag_fumante': flag_fumante})

#Cria os dados para a variavel 
dados['flag_fumante'] = dados['flag_fumante'].map({True:'Fumante', False : 'Não Fumante'})

dados.sample(5)
```

```python
#Style - Seaborn e Matplotlib
sea.set(style = 'ticks')

#lmplot
sea.lmplot( x = 'altura',
           y = 'peso',
           data = dados,
           hue= 'flag_fumante',
           palette=['tab:blue', 'tab:orange'],
           height=7 )

#Labels e título
plt.xlabel('Altura (cm)')
plt.ylabel('Peso (kg)')
plt.title('Relação Entre Altura e Peso de Fumantes e Não Fumantes')

#Remove as bordas
sea.despine()

#Show
plt.show()
```
