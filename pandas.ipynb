```python
moda = dsa_df['Quantidade'].value_counts().index[0] #moda em estastistica é uma medida tendencia central que representa o valor mais frequente em um conjunto de dados
moda#útil quando queremos saber qual é o valor mais comum ou popular em um conjunto de dados
```

```python
#por fim, preenchemos os valores NA com a moda
dsa_df['Quantidade'].fillna(value=moda, inplace=True)
dsa_df
```
